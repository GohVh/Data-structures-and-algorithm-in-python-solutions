```python
# 6.34, 6.35 skipped
# 6.36
"""
When a share of common stock of some company is sold, the capital
gain (or, sometimes, loss) is the difference between the share’s selling
price and the price originally paid to buy it. This rule is easy to understand
for a single share, but if we sell multiple shares of stock bought
over a long period of time, then we must identify the shares actually being
sold. A standard accounting principle for identifying which shares of
a stock were sold in such a case is to use a FIFO protocol—the shares
sold are the ones that have been held the longest (indeed, this is the default
method built into several personal finance software packages). For
example, suppose we buy 100 shares at $20 each on day 1, 20 shares at
$24 on day 2, 200 shares at $36 on day 3, and then sell 150 shares on day
4 at $30 each. Then applying the FIFO protocol means that of the 150
shares sold, 100 were bought on day 1, 20 were bought on day 2, and 30
were bought on day 3. The capital gain in this case would therefore be
100 · 10+20 · 6+30 · (−6), or $940. Write a program that takes as input
a sequence of transactions of the form “buy x share(s) at y each”
or “sell x share(s) at y each,” assuming that the transactions occur
on consecutive days and the values x and y are integers. Given this
input sequence, the output should be the total capital gain (or loss
) for the
entire sequence, using the FIFO protocol to identify shares
"""
class Empty(Exception):pass
class InputError(Exception):pass

class StockGainCalculator():
    def __init__(self):
        self.log = ArrayQueueRotate()
        self.log._maxlen=100000 #child class from ArrayQueue with maxlen
        self._shares = 0

    def __call__(self):
        if self._shares==0:
            command1=f"Currently you dont have any shares. Key in 'y' to purchase or 'n' to reject.\n"
            input_a = self._zerosharewill(userinput = input(command1))
            if input_a=='y': self.buy_share()
            elif input_a=='n': print('Thank you.')

        else:
            command2=f"Please key in '0' to buy shares, '1' to sell shares\n"
            input_b = self._buysellwill(userinput = input(command2))
            if int(input_b)==0: self.buy_share()
            elif int(input_b)==1: self.sell_share()

    def buy_share(self):
        command1=f"'buy x share(s) at y each'. Please key in 2 values, x and y with with space in between.\n"
        msg = input(command1)
        if not len(msg.split(" "))==2: raise InputError("Review the input, must be 'int' space 'int")
        
        sharenum, shareprice = msg.split(" ")
        sharenum, shareprice = int(sharenum), int(shareprice)
        self.log.enqueue((sharenum,shareprice))
        self._shares+=sharenum
        for i in range(self.log._n):
            print(f"Share{i+1}: Bought share={self.log[i][0]}, Bought price per share={self.log[i][1]}")

    def sell_share(self):
        if self.log.is_empty(): raise Empty('You have 0 shares.')

        command2=f"'sell x share(s) at y each'. Please key in 2 values, x and y with with space in between.\n"
        msg = input(command2)
        if not len(msg.split(" "))==2: raise InputError("Review the input, must be 'int' space 'int")

        soldnum, soldprice = msg.split(" ")
        soldnum, soldprice = int(soldnum), int(soldprice)
        if not (type(soldnum)==int and type(soldprice)==int): raise TypeError('Make sure x and y are positive integer.')
        if not (soldnum>0 and soldprice>0): raise TypeError('Make sure x and y are positive integer.')
        
        if soldnum>self._shares: raise ValueError(f'You only have {self._shares} shares left to sell.')

        inprice = 0
        outgain = soldnum*soldprice

        while soldnum>0:
            boughtnum, boughtprice = self.log.first()
            if boughtnum<=soldnum:
                self.log.dequeue()
            else:
                self.log[self.log._front]=(boughtnum-soldnum, boughtprice)
                boughtnum=soldnum

            inprice+=(boughtnum*boughtprice)
            soldnum-=boughtnum
        
        capitalgain = outgain-inprice
        print(f'Your capital gain is {capitalgain}')

    def usercheck(func):
        def inner(*args, **kwargs):
            status=True
            while status:
                will = kwargs['userinput']
                if func(will):
                    status=False
            return will
        return inner
    
    @usercheck
    def _zerosharewill(s):
        return bool((s.lower()=='y') or (s.lower()=='n'))
    
    @usercheck
    def _buysellwill(s):
        return bool(0<=int(s)<=1)
```
