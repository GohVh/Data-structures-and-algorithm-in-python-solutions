```python
# 2.4
class Flower:
    def __init__(self, name: str, petals: int, price: float):
        self._name = self._petals = self._price = None

        self.setname(name)
        self.setpetals(petals)
        self.setprice(price)

    def setname(self, name: str):
        try:
            assert type(name) == str, f"name has to be string, not {type(name)}"
            self._name = name
        except Exception as err:
            print(err)
    
    def setprice(self, price: float):
        try:
            assert type(price) == float or type(price) == int, f"price has to be float or int, not {type(price)}"
            self._price = price
        except Exception as err:
            print(err)

    def setpetals(self, petals: int):
        try:
            assert type(petals) == int, f"petals has to be int, not {type(petals)}"
            self._petals = petals
        except Exception as err:
            print(err)

    def getname(self):
        print("Flower's name has not been set.") if self._name is None else print(f'the flower is {self._name}')
    
    def getpetals(self):
        print("Number of petals has not been set") if self._petals is None else print(f'{self._name} has {self._petals} petals')
    
    def getprice(self):
        print("The price has not been set.") if self._price is None else print(f'{self._name} is RM{self._price}')
```

```python
# 2.5 & 2.6 & 2.7
import numbers

class CreditCard:
    """A consumer credit card"""

    def __init__(self, customer, bank, acnt, limit):
        self._customer = customer
        self._bank = bank
        self._acnt = acnt
        self._limit = limit
        self._balance = 0
    
    def _get_customer(self):
        return self._customer
    
    def _get_bank(self):
        return self._bank

    def _get_account(self):
        return self._acnt
    
    def _get_limit(self):
        return self._limit
    
    def _get_balance(self):
        return self._balance
    
    def charge(self, price):
        assert price != None and (type(price) == float or type(price) == int), f'price has to be in numerical int or float, not {type(price)}'
        if price + self._balance > self._limit:
            return False
        else: 
            self._balance += price 
            return True

    def make_payment(self, amount):
        try:
            assert isinstance(amount, numbers.Number)
        except AssertionError:
            print('amount must be a number, int or float')
            return
        if amount<=0:
            raise ValueError("Amount must be positive number.")
        self._balance -= amount


class CreditCardWithBalance(CreditCard):
    def __init__(self, customer, bank, acnt, limit, balance=0):
        super().__init__(customer,bank,acnt,limit)
        self._balance = balance
```

```python
# 2.9, 2.10, 2.11, 2.12, 2.13, 2.14, 2.15
class Vector:
    """Represent a vector in a multidimensional space."""
    
    def __init__(self, dim_or_n):
        """create d-dimensional vector of zeros."""
        if isinstance(dim_or_n, int):
            self._data = [0]*dim_or_n
        elif isinstance(dim_or_n, list):
            self._data = dim_or_n
        else:
            raise ValueError('Input must be dimension in int or list')
    
    def __len__(self):
        """return the dimension of the vector"""
        return len(self._data)
    
    def __getitem__(self, j):
        """return jth coordinate of vector"""
        return self._data[j]
    
    def __setitem__(self, j, val):
        """set jth coordinate of vector to given value."""
        self._data[j] = val
    
    def __eq__(self, other):
        """return true if vector has same coordinate as other"""
        return self._data == other._data

    def __ne__(self, other):
        """return true if vector differs from other"""
        return not self == other
    
    def __str__(self):
        """produce string representation of vector"""
        return '<' + str(self._data)[1:-1] + '>'
    
    def __add__(self, other):
        """return sum of two vectors."""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j]=self[j]+other[j]
        return result

    def __radd__(self, other):
        """return sum of two vectors."""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j]=other[j]+self[j]
        return result

    def __sub__(self, other):
        """return subtration of two vectors"""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j]=self[j]-other[j]
        return result

    def __rsub__(self, other):
        """return subtration of two vectors"""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')
        result = Vector(len(self))
        for j in range(len(self)):
            result[j]=other[j]-self[j]
        return result
    
    def __mul__(self, factor):
        """return multiplication product"""
        if isinstance(factor, Vector):
            result=0
            for j in range(len(self)):
                result+=self[j]*factor[j]
        else:
            result = Vector(len(self))
            for j in range(len(self)):
                result[j]=self[j]*factor
        return result

    def __rmul__(self, factor):
        """return multiplication product"""
        if isinstance(factor, Vector):
            result=0
            for j in range(len(self)):
                result+=factor[j]*self[j]
        else:
            result = Vector(len(self))
            for j in range(len(self)):
                result[j]=factor*self[j]
        return result

    def __neg__(self):
        """return negated vectors"""
        result = Vector(len(self))
        for j in range(len(self)):
            result[j]*=-1
        return result
```

```python
# 2.18
class Progression:
    """iterator producing a generic progression."""
    def __init__(self, start=0):
        """initialize current to the first value of the progression."""
        self._current=start
    
    def _advance(self):
        """update self._current to a new value.
        should be override by a subclass to customize progression."""
        self._current+=1

    def __next__(self):
        """return the next element, or raise StopIteration error."""
        if self._current is None:
            raise StopIteration()
        else:
            answer = self._current
            self._advance()
            return answer
    
    def __iter__(self):
        """an iterator must return itself as an iterator"""
        return self

    def print_progression(self,n):
        """print next n values of the progression"""
        print(" ".join(str(next(self)) for j in range (n)))

class FibonacciProgression(Progression):
    """iterator producing a generalized FIbonacci progression"""
    def __init__(self, first=0, second=1):
        super().__init__(first)
        self._prev = second-first

    def _advance(self):
        self._prev, self._current = self._current, self._prev + self._current

class ArithmeticProgression(Progression):
    def __init__(self, increment=1, start=0):
        super().__init__(start=start)
        self._increment=increment
    
    def _advance(self):
        self._current += self._increment
```

```python
# 2.22, 2.23
class Sequence:

    def __init__(self, sequence):
        self._seq = sequence
        self._k = -1
    
    def __len__(self):
        return len(self._seq)
    
    def __getitem__(self,k):
        return self._seq[k]

    def __lt__(self, i,j):
        return i<j

    def __next__(self):
        self._k +=1
        if self._k < len(self._seq):
            return(self._seq[self._k])
        else:
            raise StopIteration()

    def __iter__(self):
        return self
    
    def __eq__(self, other):
        # print(len(self._seq),len(other))
        assert len(self._seq)==len(other), f'Dimension of both sequences has to be same'
        for i in range(len(self._seq)):
            if self._seq[i] == other[i]:
                print("True")
            else: print(False)
```

```python
# 2.27
class Range:
    def __init__(self, start, stop=None, step=1):
        if step==0:
            raise ValueError('step cannot be 0')
        
        if stop is None:
            start, stop = 0, start
        
        self._length = max(0, (stop-start+step-1)//step)
        self._start = start
        self._step = step
    
    def __len__(self):
        return self._length
    
    def __getitem__(self, k):
        if k < 0:
            k += len(self)
        
        if not 0 <= k < self._length:
            raise IndexError("index out of range")
        
        return self._start + k * self._step

    def __contain__(self, x):
        if not self._start <= x < self._stop:
            return False
        if (x - self._start) % self._step == 0:
            return True
        else:
            return False
```

```python
# 2.34
class ReadDoc:
    def __init__(self, file):
        self._doc = open(file,"r").read()
        self._charlist = self._createdict('a')
        self._capcharlist = self._createdict('A')
        

    def _readfile(self):
        for i in self._doc:
            if (ord('a')<=ord(i)<=ord('a')+26):
                self._charlist[i]+=1
            elif (ord('A')<=ord(i)<=ord('A')+26):
                self._capcharlist[i]+=1
    
    def _plotchart(self):
        for i in self._charlist:
            print(f"{i}: {'x'*self._charlist[i]}")
        for i in self._capcharlist:
            print(f"{i}: {'x'*self._capcharlist[i]}")

    def _createdict(self,start):
        return {chr(i):0 for i in range(ord(start), ord(start)+26)}
```

```python
# 2.35
import random
class character:
    def __init__(self, object, partner):
        self.CHANCE_OF_PACKING = 0.5
        self._object = object
        self._partner = partner
    
    def generate_packets(self):
        if random.random()<=self.CHANCE_OF_PACKING:
            self._packets=[chr(random.randint(ord('a'), ord('a')+25)) for i in range(4)]
            print(f"{self._object} generated packets {self._packets} for {self._partner}")
            return self._packets
        else: return False
        
    def send_packets(self, packets):
        self._readypackets = packets
        if self._readypackets is not False:
            print(f'{self._object} is sending {self._readypackets} to {self._partner}')
            return self._readypackets
        else: return False

    def receive_packets(self, packets):
        self._receivedpackets = packets
        if self._receivedpackets is not False:
            print(f'{self._object} has read {self._receivedpackets} from {self._partner} and is deleting it.')
```

```python
class RiverEco:

    MOVE_CHANCE = 0.2
    LEFT_RIGHT = 0.5

    def __init__(self, totalnum, bearnum=2, fishnum=10):
        
        self._totalnum=totalnum
        self._bearnum=bearnum
        self._fishnum=fishnum
        self._ecolist = [None for i in range(totalnum)]

    def _init_assign_animal(self):
        randlist = random.sample(range(0,self._totalnum-1),self._bearnum+self._fishnum)
        bearlist = random.sample(randlist, self._bearnum)
        fishlist = randlist.copy()
        for i in randlist:
            if i in bearlist:
                fishlist.remove(i)
        
        for i in bearlist:
            self._ecolist[i]=Bear('B',i, self._ecolist)
        for i in fishlist:
            self._ecolist[i]=Fish('F',i, self._ecolist)
    
    def check_species(self, location1, location2):
        animal1, animal2 = self._ecolist[location1], self._ecolist[location2]
        if animal1.getname() == animal2.getname(): return True # return True = create new
        else: return False # return False = eat prey (eat Fish)
    
    def startgame(self):
        self._init_assign_animal()
        if random.random()>self.MOVE_CHANCE:
            for i in range(len(self._ecolist)):
                if self._ecolist[i] != None:
                    if random.random()>self.LEFT_RIGHT and i+1<len(self._ecolist):
                        if self._ecolist[i+1] != None:
                            if self.check_species(i, i+1):
                                self.createnew()
                            else: self.remove(i,i+1)
                        else:
                            self._ecolist[i+1]=self._ecolist[i]
                            self._ecolist[i] = None
                    elif random.random()<=self.LEFT_RIGHT and i-1>=0:
                        if self._ecolist[i-1] != None:
                            if self.check_species(i, i-1):
                                self.createnew()
                            else: self.remove(i,i-1)
                        else:
                            self._ecolist[i-1]=self._ecolist[i]
                            self._ecolist[i] = None
        self.checkeco()
    
    def checkeco(self):
        for i in range(1,len(self._ecolist)-1):
            if self._ecolist[i] != None:
                if self._ecolist[i-1] != None:
                    if not self.check_species(i, i-1):
                        self.remove(i,i-1)
                elif self._ecolist[i+1] != None:
                    if not self.check_species(i, i+1):
                        self.remove(i,i+1)
    
    def createnew(self):
        r = random.randint(0,len(self._ecolist)-1)
        while self._ecolist[r]==None:
            r = random.randint(0,len(self._ecolist)-1)
        name = self._ecolist[r].getname()
        if name == 'B':
            self._ecolist[i]=Bear('B', r, self._ecolist)
        elif name == 'F':
            self._ecolist[i]=Bear('F', r, self._ecolist)
    
    def remove(self, location1, location2):
        animal1, animal2 = self._ecolist[location1], self._ecolist[location2]
        if animal1.getname() == 'B':
            self._ecolist[location2] = None
        elif animal2.getname() == 'B': 
            self._ecolist[location1] = None
    
    def printgame(self):
        eco=''
        for i in self._ecolist:
            if i != None:
                eco+=i.getname()
            else: eco+='-'
        print(eco)
        
```

```python
# 2.38, exclude expired copyrights and real book population
# readbook method can substitued with python built-in function, open(fileurl) to read file
# add one more parameter -> bookurl to booklist dict
class ebookReader:
    def __init__(self, user, bookshoplist, userbooklist):
        self._name = user
        self._shop = BookShop(bookshoplist)
        self._user = PurchasedBook(userbooklist)

    def readbook(self, book=None):
        userlist = self._user._returnlist()
        if book is not None:
            assert book in userlist, print('Book requested not available, please purchase in shop')
            print(f"reading {book}...")
        else:
            print(f"Purchased Book List: ----------------")
            self._user._viewlist()
            book = str(input('Select a book you want to read'))
            assert book in userlist, print('Book requested not available, please purchase in shop')
            print(f"reading {book}...")

    def purchasebook(self):
        shoplist = self._shop._returnlist()
        print('Book Shop Book List: ----------------')
        self._shop._viewlist()
        desiredbook = str(input('Please key in the name of the book wanted to purchase'))
        assert desiredbook in shoplist, print('Book requested not available')
        self._user._addbook(desiredbook, shoplist[desiredbook][0], True)
        self._shop._removebook(desiredbook)
    
    def showpurchased(self):
        self._user._viewlist()

class BookList:
    def __init__(self, booklist):
        if booklist == None:
            self._booklist = {}
        self._booklist = booklist

    def _addbook(self, book, price, status):
        self._book = book
        self._price = price
        self._status = status
        self._booklist[self._book]=[self._price, self._status]

    def _removebook(self, book):
        self._booklist.pop(book)

    def _viewlist(self):
        for i in self._booklist:
            print(f"{i}, price: {self._booklist[i][0]}, Purchased? {self._booklist[i][1]}")

    def _returnlist(self):
        return self._booklist

class PurchasedBook(BookList):
    def __init__(self, booklist):
        super().__init__(booklist)

class BookShop(BookList):
    def __init__(self, booklist):
        super().__init__(booklist)
```
